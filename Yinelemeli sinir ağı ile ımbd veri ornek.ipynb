```python
from keras.datasets import  imdb
from keras.preprocessing import sequence
```

```python
num_features=1000
maxlen=500
batch_size=32
```

```python
print("load data.....")
(input_train,y_train),(input_test,y_test)=imdb.load_data(num_words=num_features)


print(len(input_train),"eğitim dizisi",input_train.shape)
print(len(input_test),"test dizisi",input_test.shape)
```

```python

print('Pad sequnce (sample x train)')

input_train = sequence.pad_sequences(input_train, maxlen=maxlen)
input_test = sequence.pad_sequences(input_test, maxlen=maxlen)


print(len(input_train), 'Eğitim dizisi', input_train.shape)
print(len(input_test), 'test dizisi', input_test.shape)
```

```python
from keras.layers import Dense
from keras import layers
```

# basit rnn ile modelleme

```python
model=Sequential()
model.add(Embedding(num_features,32))
model.add(SimpleRNN(32))
model.add(Dense(1,activation="sigmoid"))
```

# basit bir lsth ile modelleme


```python

model = Sequential()
model.add(layers.Embedding(num_features, 32))
model.add(layers.Bidirectional(layers.LSTM(32)))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])

history = model.fit(input_train, y_train,
                    epochs = 10,
                    batch_size = 128,
                    validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
acc=history.history["acc"]
val_acc=history.history["val_acc"]
loss=history.history["loss"]
val_loss=history.history["val_loss"]
```

```python
epochs=range(1,len(acc)+1)
```

```python
plt.plot(epochs,acc,"m*-",label="egitim başarisi ")
plt.plot(epochs,val_acc,"g-",label="dogrulama başarisi ")
plt.title("başarı")
plt.legend()
plt.figure()
plt.plot(epochs,loss,"m*-",label="egitim kaybı ")
plt.plot(epochs,val_loss,"g-",label="dogrulama kaybı ")
plt.title("kayıp")
plt.legend()


plt.show()
```
