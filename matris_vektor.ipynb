# işlemler
*eleman temelli*
bu işlemler tensörün her elemanına ayrı ayrı uygulanır

```python
def navie_add(x,y):
  assert len(x.shape)==2 # x ve y 2 boyutlu numpy tensörü
  assert x.shape==y.shape

  x=x.copy()

  for i  in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j]+=y[i,j]
  return x
```

```python
x=np.array([[7,14,21],
 [35,42,49],
  [56,63,70]])
y=np.array([[1,4,51],
 [5,2,9],
  [56,63,70]])
z=navie_add(x,y)
z
```

```python
def navie_relu(x):
  assert len(x.shape)==2

  x=x.copy()

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
     x[i,j]=max(x[i,j],0)
  return x
```

```python
x=np.array([[7,14,21],
 [35,42,49],
  [56,63,70]])
y=np.array([[1,-4,-51],
 [5,-2,-9],
  [56,63,70]])

z=navie_relu(x)# maks değeri

z
```

# Yayma operatörü

```python
def navie_add_matrix_and_vector(x,y):
  assert len(x.shape)==2
  assert len(y.shape)==1
  assert x.shape[1]==y.shape[0]


  x=x.copy()

  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i,j]+=y[j]
  return x

```

```python
x=np.array([[7,14,21],
 [35,42,49],
  [56,63,70]])
y=np.array([1,2,3])
z=navie_add_matrix_and_vector(x,y
)
z
```
